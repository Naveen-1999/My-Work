# Calculating VIF

```python
#'Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight'

rsq_Age_08_04 = smf.ols('Age_08_04~KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data_1).fit().rsquared  
vif_Age_08_04 = 1/(1-rsq_Age_08_04) 

rsq_KM = smf.ols('KM~Age_08_04+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data_1).fit().rsquared  
vif_KM = 1/(1-rsq_KM)

rsq_HP = smf.ols('HP~Age_08_04+KM+cc+Doors+Gears+Quarterly_Tax+Weight',data=data_1).fit().rsquared  
vif_HP = 1/(1-rsq_HP) 

rsq_cc = smf.ols('cc~Age_08_04+KM+HP+Doors+Gears+Quarterly_Tax+Weight',data=data_1).fit().rsquared  
vif_cc = 1/(1-rsq_cc) 

rsq_DO = smf.ols('Doors~Age_08_04+KM+HP+cc+Gears+Quarterly_Tax+Weight',data=data_1).fit().rsquared  
vif_DO = 1/(1-rsq_DO)

rsq_GE = smf.ols('Gears~Age_08_04+KM+HP+cc+Doors+Quarterly_Tax+Weight',data=data_1).fit().rsquared  
vif_GE = 1/(1-rsq_GE) 

rsq_QT = smf.ols('Quarterly_Tax~Age_08_04+KM+HP+cc+Gears+Weight+Doors',data=data_1).fit().rsquared  
vif_QT = 1/(1-rsq_QT)

rsq_WE = smf.ols('Weight~Age_08_04+KM+HP+cc+Doors+Quarterly_Tax+Gears',data=data_1).fit().rsquared  
vif_WE = 1/(1-rsq_WE) 


# Storing vif values in a data frame
d1 = {'Variables':['Age_08_04','KM','HP','cc','Doors','Gears','Quarterly_Tax','Weight'],'VIF':[vif_Age_08_04,vif_KM,vif_HP,vif_cc,vif_DO,vif_GE,vif_QT,vif_WE]}

Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

```python
# All vif values are less than 10
```

# Residual Analysis

# Test for Normality of Residuals(Q-Q Plot)

```python
import statsmodels.api as sm
qqplot=sm.qqplot(Model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(Model.resid>4000))   #outliar detection from above QQ plot of residuals
```

```python
list(np.where(Model.resid<-4000))
```

# Test for Homoscedasticity or Heteroscedasticity
(plotting model's standardized fitted values vs standardized residual values)



```python
def standard_values(vals) : return (vals-vals.mean())/vals.std()  # User defined z = (x - mu)/sigma
```

```python
plt.scatter(standard_values(Model.fittedvalues),standard_values(Model.resid))
plt.title('Residual Plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show() 

```

# Model Deletion Diagnostics
# Detecting Influencers/Outliers using Cook's distance¶

```python
Model_influence = Model.get_influence()
(c,_)=Model_influence.cooks_distance
```

```python
c               #1. Cook's Distance: If Cook's distance > 1, then it's an outlier
```

```python
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(data_1)),np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```
