# Collebrative Filtering

## UBCF with correlation matrix as metric

```python
ubcf_data=pd.pivot_table( data=movie_data, values='rating', index='movie',columns='userId').fillna(0)
ubcf_data.columns = movie_data.userId.unique()
ubcf_data
```

```python
ubcf_data.corr().round(3)
```

# UBCF  by using eucleideon  distance as a metric

```python
from sklearn.metrics import pairwise_distances
```

```python
ubcf_data_ecd = pd.pivot_table(data = movie_data,values='rating',index='userId',columns='movie').fillna(0)
ubcf_data_ecd.index = movie_data.userId.unique()
ubcf_data_ecd
```

```python
euclideon_ubcf = pairwise_distances(X = ubcf_data_ecd,metric='euclidean')
euclideon_ubcf_df = pd.DataFrame(data = euclideon_ubcf)
euclideon_ubcf_df.index = movie_data.userId.unique()
euclideon_ubcf_df.columns = movie_data.userId.unique()
euclideon_ubcf_df
```

 # UBCF  by using Cosine  distance as a metric

```python
ubcf_data_cosine = pd.pivot_table(data = movie_data,values='rating',index='userId',columns='movie').fillna(0)
ubcf_data_cosine.index = movie_data.userId.unique()
ubcf_data_cosine
```

```python
cosine_distance = 1-pairwise_distances(X = ubcf_data_cosine,metric='cosine')
cosine_distance_df = pd.DataFrame(data = cosine_distance)
cosine_distance_df.index = movie_data.userId.unique()
cosine_distance_df.columns = movie_data.userId.unique()
cosine_distance_df
```
