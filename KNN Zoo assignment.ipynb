```python
# How to find Optimal Number Of Clusters
```

```python
from sklearn.model_selection import cross_val_score


container_cv_scores = []

for i in range(1,30):
    knn_model= KNeighborsClassifier(n_neighbors=i)
    cv_scores= cross_val_score(estimator=knn_model,X=X_scaled,y=y,cv=5)
    container_cv_scores.append(cv_scores.mean())
   
```

```python
container_cv_scores
```

```python
plt.plot(range(1,30),container_cv_scores)
```

```python
container_cv_scores.index(max(container_cv_scores)) 
```

**k=7 is the optimal number**
