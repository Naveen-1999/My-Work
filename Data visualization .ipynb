```python
import pandas as pd
```

```python
data=pd.read_csv('C:\\Users\\user\\Downloads\\mtcars.csv')
data.head()
```

```python
data['mpg'].plot(kind="kde")
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data['mpg']
```

```python
plt.figure( figsize=(7,7))
plt.hist(x='mpg',data=data)
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.show()
```

```python
import seaborn as sns
```

```python
data['carb'].value_counts()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='carb',data=data)
plt.show()
```

```python
data['gear'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(data['gear'].value_counts(),labels=[3,4,5],autopct='%1.0f%%', labeldistance=1.2,explode=[0.02,0.02,0],shadow=True)
plt.show()
```

```python
plt.boxplot(x='mpg',data=data)
plt.show()
```

```python
plt.scatter(x='hp',y='mpg',data=data)
plt.xlabel('hp')
plt.ylabel('mpg')
plt.show()
```

# Bar Plot - x should be discrete and y =continuous

```python
data.head()
```

```python
sns.barplot(x='cyl',y='mpg',data=data)
plt.show()
```
