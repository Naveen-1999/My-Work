# Model Building

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

# Lets do Association Rule by giving 
## Support =10%
## Confidence =70%

```python
frequent_items =apriori(df=book_data,min_support=0.1,use_colnames=True)
frequent_items
```

```python
best_associates =association_rules(df=frequent_items, metric='lift',min_threshold=0.7)
best_associates
```

support(A->C) = support(A+C), range: [0, 1]

confidence(A->C) = support(A+C) / support(A), range: [0, 1]

lift(A->C) = confidence(A->C) / support(C), range: [0, inf]

leverage(A->C) = support(A->C) - support(A)*support(C),
    range: [-1, 1]

conviction = [1 - support(C)] / [1 - confidence(A->C)],
    range: [0, inf]

```python
# leverage values of 0 indicates the indepence so it ranges between [-1 1]
```

```python
best_associates.sort_values('lift',ascending=False)
```

```python
best_associates[best_associates.lift>1]
```

```python
#  lift>1 is a good influential rule in selecting the associated transcations.
```
