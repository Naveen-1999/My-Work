# Model Building

```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols('Profit~RDS+ADMS+MS', data=Data1).fit()
```

```python
model.summary() # model perfomance
```

```python
model.params # both slopes and intercept
```

```python
print('R2 Score    : ',round(model.rsquared,4)) #The Contribution of 1 independent feature for prediction.
print('Adj.R2 Score: ',round(model.rsquared_adj,4))
print('AIC Value   : ',round(model.aic,4))
print('BIC Value   : ',round(model.bic,4))
```

```python
# Model accuracy is 94.75%
```

# Hypothesis Formulation

```python
print(model.tvalues, '\n' , model.pvalues)
```

```python
model_1=smf.ols(formula = 'Profit ~ ADMS', data = Data1).fit()
model_1
```

# Test Data

```python
y_pred_test=linear_model.predict(x_test)
mean_squared_error(y_test,y_pred_test)
```

# Residual Analysis


# Test for Normality of Residuals(Q-Q Plot)

```python
import statsmodels.api as sm
sm.qqplot(model.resid,line='q')
plt.title('Normal Q-Q plot residuals')
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

# Homoscedascity Test

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

# Model Deletion Diagnostics


# Detecting influencers/outliers using cooks distance

```python
model_influence=model.get_influence()
(c,_) = model_influence.cooks_distance
c
```

```python
 #Plot the influencers using the stem plot
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(data_50)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```
